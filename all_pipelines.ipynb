```python
from nilearn import datasets
import pandas as pd
import numpy as np
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.svm import LinearSVC
import prepare_data

```

```python
data_dir = f'D:/fer/projekt_r/data_'
pipelines = ['cpac', 'ccs', 'niak', 'dparsf']
```

```python
all_accuracies = {'cpac': [], 'ccs': [], 'niak': [], 'dparsf': []}

for pipeline in pipelines:
    data = ''
    data = data_dir + pipeline
    output = data

    abide = datasets.fetch_abide_pcp(data_dir = data,
                                     pipeline = pipeline, 
                                     quality_checked = True,
                                     legacy_format = False)

    abide_pheno = pd.DataFrame(abide.phenotypic)

    groups = []
    for site in abide_pheno.SITE_ID:
        groups.append(site)

    X, y = prepare_data.prepare_data(data_dir = data,
                                     output_dir = output,
                                     pipeline = pipeline)

    y = y.values
    
    logo = LeaveOneGroupOut()
    logo.get_n_splits(X, y, groups)

    l_svc = LinearSVC(max_iter = 10000)

    for i, (train_index, test_index) in enumerate(logo.split(X, y, groups)):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        print(f'Training model {i + 1} for {pipeline} pipeline.')
        l_svc.fit(X_train, y_train)
        all_accuracies[pipeline].append(l_svc.score(X_test, y_test))

    print(f'Finished training for {pipeline} pipeline.')
    print(f'==============================================')



for pipeline, accuracies in all_accuracies.items():
    print(f'Average accuracy for {pipeline} pipeline: {np.mean(accuracies)}')
    print(f'Maximum accuracy for {pipeline} pipeline: {np.max(accuracies)}')
    print(f'Minimum accuracy for {pipeline} pipeline: {np.min(accuracies)}')
    print(f'===============================================================')
```

```python
accuracies_df = pd.DataFrame(all_accuracies)

summary = accuracies_df.agg(['mean', 'max', 'min'])
summary = summary.sort_values(by = 'mean', axis = 1, ascending = False)
summary
```
